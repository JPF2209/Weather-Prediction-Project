```python
import pandas as pd
from datetime import datetime, timedelta
from pytz import timezone
```

```python
weather = pd.read_csv("CSV.csv", index_col = "Humidity")
weather.drop(weather.tail(1).index,inplace=True)
```

```python
weather['Date'] = pd.to_datetime(weather['Date'], unit='s')
weather['Date'] = weather['Date'].dt.strftime('%H:%M:%S')
weather = weather[weather['Date'] < "07:09:00"]
weather
```

```python
weather['Time'] = pd.to_datetime(weather['Time'], unit='s')
weather['Humid'] = weather.index
weather['T_Target'] = weather.shift(-1)['Temperature']
weather['H_Target'] = weather.shift(-1)['Humid']
weather['L_Target'] = weather.shift(-1)['Light Intensity']
weather.columns = weather.columns.str.strip()
weather
```

```python
weather["T_Delta"] = weather["T_Target"] - weather["Temperature"]
weather["H_Delta"] = weather["H_Target"] - weather["Humid"]
weather["L_Delta"] = weather["L_Target"] - weather["Light Intensity"]
weather = weather.sort_values(['Date'], ascending = True)
weather
```

```python
weather = weather.drop(["L_Target"], axis = 1)
weather = weather.drop(["H_Target"], axis = 1)
weather = weather.drop(["T_Target"], axis = 1)
weather
```

```python
i = 0
mean_all = [[] for i in range(6)]
verify = False
prev_h = 2
new_index = 0
for index, row in weather.iterrows():
    t = row['Time']
    if prev_h != t.hour:
        verify = False
    if t.minute == 30 and verify == False:
        verify = True
        new_index += 1
    prev_h = t.hour
    mean_all[new_index].append(row)

mean_all
```

```python
j = -1
first = True
weather_alt = weather
k = j + 1

for i in range(len(mean_all)):  
    val = len(mean_all[i])  
    l = val + j
    if first == True:
        first = False        
        hour = weather_alt.iloc[k:l]
        j = k + val
    else:        
        hour = weather_alt.iloc[j:l]
        j += val
    hour.dropna()
    hour.to_csv(f"hour{i + 2}.txt", index = False)
```

```python
import numpy as np
from sklearn import preprocessing, svm

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

def calculate_temp(temperature, humidity, light_intensity, hour, delta):
    data = pd.read_csv(f"hour{hour}.txt", delimiter=",", low_memory=False)
    print(delta)
    x = np.array(data[["Humid", "Light Intensity", "Temperature"]])
    y = np.array(data[[delta]])
 
    model = LinearRegression().fit(x,y)

    r_sq = model.score(x, y)
    print(f"coefficient of determination: {r_sq}")

    arr = np.array([[humidity, light_intensity, temperature]], dtype=object)

    delta_temp = model.predict(arr)
    delta_temp = delta_temp[0][0]
    temperature = temperature + delta_temp
    return temperature
```

```python
hours = []
hour = 2
#Input goes here
input_temperature = 23
light_intensity = 500
time = 16
wanted_hour = 14
humidity = 67
input_hour = (time - 12) + hour
utcOffset = 10
temp = []

for i in range(len(mean_all)):
    if input_hour != hour:
        temperature = calculate_temp(input_temperature, humidity, light_intensity, hour, "T_Delta")
    else:
        temperature = input_temperature

    temp.append(temperature)
    if (hour > 24): hour -= 24
    
    hour_str = str(hour+utcOffset) if hour+utcOffset < 24 else str(hour + utcOffset - 24)
    hours.append(hour_str)
    hour += 1

# print(hours)
j = 0
final_hour = wanted_hour - 12
final_temp = temp[final_hour]
print(final_temp)
# final_temp = 
```
